**Задание 1**

Функция: $f(x) = sin(\sqrt{2x})$

Отрезок: $[a=0,b=1]$

Шаг разбиения: $dx=0,01$

```python
import numpy
import matplotlib.pyplot as plt
```

```python
def f(x):
    return numpy.sin(numpy.sqrt(2*x))
```

```python
a=0
b=1
dx=1e-2

x_1 = numpy.arange(a,b-dx,dx)

f_values = numpy.array(f(x_1))

```

**Задание 2**

Построение графика функции $f(x) = sin(\sqrt{2x})$

```python
plt.plot(x_1,f_values,label='$sin(\sqrt{2x})$')

plt.title('График функции', fontsize = 16)

plt.xlabel('x', fontsize = 12)
plt.ylabel('y',fontsize = 12)

plt.legend(fontsize = 12)

plt.show()
```

**Задание 3**

Построение точечного графика функции $f(x) = sin(\sqrt{2x})$

```python
plt.scatter(x_1,f_values,marker='.', s=30,c=(0.5,0.5,0.9),label='$sin(\sqrt{2x})$')

plt.title('Точечный график функции')

plt.xlabel('x', fontsize = 12)
plt.ylabel('y', fontsize = 12)

plt.grid(True, color='green', linestyle='--', alpha=0.70)
plt.legend(fontsize=10)

plt.show()

```

**Задание 6**

Построить трехмерный график функции $(x+3)^3+(y-3)^2$, где $0<=x<=10$ и $0<=y<=10$


```python
def function_3d(x,y):

    return (x+3)**3+(y-3)**2
```

```python
from mpl_toolkits.mplot3d import Axes3D

x_range = (0,10)

y_range = (0,10)

dy =dx

x= numpy.arange(x_range[0],x_range[1]-dx, dx)

y= numpy.arange(y_range[0],y_range[1]-dy, dy)

X,Y = numpy.meshgrid(x,y)

Z=function_3d(X,Y)

figur = plt.figure(figsize=(7,7))

graphic = figur.add_subplot(projection = '3d')
graphic.plot_surface(X,Y,Z,color='skyblue', label = '$(x+3)^3+(y-3)^2$')

graphic.set_xlabel('x', fontsize=12)
graphic.set_ylabel('y', fontsize=12)
graphic.set_zlabel('z', fontsize=12)

graphic.set_title('Трехмерный график функции', fontsize=14)

plt.legend(fontsize=12)



plt.show()
```

**Задание 7**

Используя подграфики построить сетку 2x2 из 4 предыдущих графиков 

```python
def build_style_graohic(style):
    fig = plt.figure(figsize=(12, 8))

    fig.suptitle('Сетка графиков', fontsize=16)

    ax1 = fig.add_subplot(2,2,1)
    ax1.plot(x_1,f_values,label='$sin(\sqrt{2x})$')
    ax1.set_title('График функции', fontsize = 16)
    ax1.set_xlabel('x', fontsize = 12)
    ax1.set_ylabel('y',fontsize = 12)
    ax1.legend(fontsize = 12)
    ax1.set_facecolor('lightblue')

    ax2 = fig.add_subplot(2,2,2)

    ax2.scatter(x_1,f_values,marker='.', s=50,c=(0.5,0.5,0.9),label='$sin(\sqrt{2x})$')
    ax2.set_title('Точечный график функции')
    ax2.set_xlabel('x', fontsize = 12)
    ax2.set_ylabel('y', fontsize = 12)
    ax2.grid(True, color='green', linestyle='--', alpha=0.45)
    ax2.legend(fontsize=10)
    ax2.set_facecolor('lightgreen')

    ax3 = fig.add_subplot(2,2,3)

    ax3.pie(counts, labels=unique,autopct='%1.1f%%', startangle=0,colors=['palegreen', 'blue', 'lightcoral', 'gold'])
    ax3.set_title('Круговая диаграмма равномерной выборки', fontsize = 12)

    ax4 = fig.add_subplot(2,2,4, projection='3d')  

    ax4.plot_surface(X, Y, Z, color= 'syblue' label= '$(x+3)^3+(y-3)^2$')
    ax4.set_xlabel('x', fontsize=12)
    ax4.set_ylabel('y', fontsize=12)
    ax4.set_zlabel('z', fontsize=12)
    ax4.set_title('Трехмерный график', fontsize=12)
    ax4.set_facecolor('lightyellow')
    ax4.legend(fontsize=10)

    plt.tight_layout()

    plt.show()




```

```python
plt.style.use('default')
build_style_graohic('default')
```
